# Imports

```python
import pandas as pd
import numpy as npz
import matplotlib.pyplot as plt

from sklearn.mixture import GaussianMixture
```

# Load data

```python
dataset = pd.read_csv("Wine-quality-challenge.csv")
dataset.head()
```

# clustering

optimal clusters: What are the best cluster for appling and interpretating:

```python
n_components = np.arange(1,10)
models = [GaussianMixture(n, random_state = 1502).fit(dataset)
          for n in n_components]

plt.plot(n_components,
         [m.bic(dataset) for m in models],
         label = 'BIC')
plt.plot(n_components,
         [m.aic(dataset) for m in models],
         label = 'AIC')
plt.legend(loc = 'best')
plt.xlabel('number of components')
```

- This not clear what the cluster we use;
- We can use 4, 5 and 6, that good, but use 4 because more simples to explain and interpretating.

# Gaussian Mixture Model

```python
model = GaussianMixture(n_components= 4,
                        random_state = 1502).fit(dataset)
```

# Interpretation

```python
interpretation = pd.DataFrame(model.means_,
                              columns = list(dataset.columns))
interpretation
```
